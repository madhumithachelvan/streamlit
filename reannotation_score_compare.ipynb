```python
import pandas as pd
import os
```

```python
org_annot = pd.read_csv('./reannotations/ori_annotator.csv',)
```

```python
org_annot
```

```python
re_annot = pd.read_csv('./reannotations/merged_reannotations.csv')
org_p_id = org_annot['p_id'].to_list()
org_text_id = org_annot['text_id'].to_list()
        
```

```python
for i,row in re_annot.iterrows():
    p_id = row['p_id']
    text_id = row['text_id']
    if p_id in org_p_id and text_id in org_text_id:
        org_annot.loc[(org_annot['p_id']==p_id) & (org_annot['text_id']==text_id), "new_style_score"] = row['style_score']
        org_annot.loc[(org_annot['p_id']==p_id) & (org_annot['text_id']==text_id), "new_confidence_score"] = row['confidence_score']
```

```python
org_annot.loc[:,'style_comp'] = (org_annot.loc[:,'style_score'] == org_annot.loc[:,'new_style_score'])
org_annot.loc[:,'confidence_comp'] = (org_annot.loc[:,'confidence_score'] == org_annot.loc[:,'new_confidence_score'])
```

```python
with pd.option_context('display.max_rows', None,
                       'display.max_columns', None,
                       'display.precision', 3,
                       ):
    print(org_annot.loc[:,['confidence_score','new_confidence_score','confidence_comp']].dropna())
```

```python
org_annot.to_csv('new_annotator.csv',index=False)
```
